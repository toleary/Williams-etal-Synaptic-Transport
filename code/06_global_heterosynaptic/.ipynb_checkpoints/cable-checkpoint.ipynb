**Analytical solution:**

Assume a cable of length $L$, with sealed boundary conditions. Let $u(x,t)$ be the distribution of $u$ along the cable at time $t$ and position $x$. Let $u^\star(x,t)$ be the distribution of detached/active cargo.

PDE:
$$u_t = \underbrace{f(x)u_{x}}_{\text{advection}} - \underbrace{g(x)u}_{\text{dissipation}}$$

Initial Condition:
$$u(x,0) = \left\{\begin{matrix}
1,~~x=0\\ 
0,~~x\neq0
\end{matrix}\right.$$

Concentration of active/detached cargo:
$$u^{\star}(x,t) = g(x) \int_0^t u(x,s)ds$$

Steady-state:
$$u^{\star}_{ss}(x) = g(x) \int_0^\infty u(x,s)ds$$

**Problem:** Find $f(x)$ and $g(x)$ such that $u^{\star}$ approaches a desired spatial distribution $\tilde{u}^\star(x)$

**Constrained Problem:** Find $f(x) = F$ (const.) and $g(x)$ such that: $||u^{\star}_{ss}(x) - \tilde{u}^\star(x)||_2$ is minimized

```python
from __future__ import division
import numpy as np
from scipy.integrate import odeint
from scipy.linalg import expm # matrix exponential solves linear system
import pylab as plt
%matplotlib inline

N = 100 # number of compartments

# initial condition
u0 = np.zeros(2*N)
u0[0] = 1.0

def get_matrix(a=np.ones(N-1)*75,b=np.ones(N-1),c=np.ones(N)):
    # calculate state-transition matrix
    mid = -c
    mid[:-1] -= a
    mid[1:]  -= b
    A1 = np.diag(a, -1) + np.diag(mid) + np.diag(b, 1) # Upper-right block
    A2 = np.diag(c) # Lower-left block

    return np.bmat([[A1,np.zeros((N,N))],[A2,np.zeros((N,N))]])
```

```python
A = get_matrix(a=np.ones(N-1)*10,c=np.zeros(N))

nt = 15
cols = np.array([np.linspace(0,1,nt),np.zeros(nt),np.linspace(1,0,nt)]).T

plt.figure()
plt.plot(u0[:N],'-b',lw=2)
for (i,t) in enumerate(np.logspace(-1.5,1.5,nt-1)):
    u = np.dot(expm(A*t),u0)
    plt.plot(u[:N],color=cols[i,:],lw=2)

plt.title('traveling "wave" along microtubules -- constant velocity')
plt.ylim([0,0.3])
plt.show()
```
