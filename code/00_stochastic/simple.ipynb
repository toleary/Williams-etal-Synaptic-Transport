```python
from __future__ import division
import pylab as plt
from numpy import random
import numpy as np
%matplotlib inline
```

```python
def lgc(x,half,slope,hi,lo):
    return lo + (hi-lo)/(1 + np.exp((half-x)/slope))

def gauss(x,a,b,c,d):
    return d + a*np.exp(-(x-b)**2 / (2*c**2))

def ret_v(x):
    return -lgc(x,0.5,1,0.9,0.2)

def ant_v(x):
    return lgc(x,0.5,3,1.1,0.1)

def ret_run(x):
    v = ret_v(x)*random.uniform(0.9,1.1)
    dt = random.exponential(0.3)
    return (v*dt,dt) # return run length, duration
    
def ant_run(x):
    v = ant_v(x)*random.uniform(0.9,1.1)
    dt = random.exponential(0.3)
    return (v*dt,dt) # return run length, duration

def pause_run(x):
    dt = random.exponential(0.3)
    return (0.0,dt)  # return run length, duration

def ret_prob(x):
    return lgc(x,0.3,1,2,0.3)
    
def ant_prob(x):
    return lgc(x,0.3,1,1,0.1)

def pause_prob(x):
    return gauss(x,5,1,2,0.1)

def probabilities(x):
    p = np.array([ret_prob(x), pause_prob(x), ant_prob(x)])
    return p/np.sum(p,axis=0) # normalize

def update(x,t):
    # stochastically determine next behavior (run or pause)
    run = random.choice([ret_run, pause_run, ant_run],p=probabilities(x[-1]))
    # determine change in position and time (also stochastic)
    dx,dt = run(x[-1])
    # update position and time
    x.append(x[-1]+dx)
    t.append(t[-1]+dt)

def sim(x0=0,t0=0,tmax=100):
    x,t = [x0],[t0]
    while t[-1] < tmax: update(x,t)
    return x,t
```

```python
plt.figure()
for _ in range(50):
    x,t = sim()
    plt.plot(t,x,'-k',alpha=0.5)
plt.xlabel('time')
plt.ylabel('position')
plt.show()
```

```python
x,t = sim()
dt = np.diff(t)
dx = np.diff(x)
plt.figure()
plt.plot(np.abs(dx[np.abs(dx)>0.1]/dt[np.abs(dx)>0.1]),dx[np.abs(dx)>0.1],'.k')
plt.xlabel('Run Velocity')
plt.ylabel('Run Length')
plt.show()
```

```python
xx = np.linspace(-10,10,500)
P = probabilities(xx).T
plt.figure()
plt.plot(xx,P)
plt.ylabel('Probability')
plt.xlabel('Position')
plt.legend(['ret','pause','ant'],loc='best')
plt.show()
```

```python
plt.figure()
plt.plot(xx,ret_v(xx))
plt.plot(xx,0*xx)
plt.plot(xx,ant_v(xx))
plt.ylabel('Average Velocity')
plt.xlabel('Position')
plt.legend(['ret','pause','ant'],loc='best')
plt.show()
```

```python
plt.figure()
plt.plot(xx,ret_v(xx)*P[:,0] + ant_v(xx)*P[:,2],'-k')
plt.ylabel('Expected Velocity')
plt.xlabel('Position')
plt.show()
```
