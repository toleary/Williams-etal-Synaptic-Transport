```python
import pylab as plt
from numpy import random,argwhere,linspace
import numpy as np
%matplotlib inline
```

```python
velocities = [-0.5, 0.0, 0.5]
probabilities = [0.3, 0.3, 0.4]

def update(x):
    i = argwhere(random.multinomial(1,probabilities))[0][0]
    v = velocities[i]
    x.append(x[-1]+v*dt) # append new position
```

```python
dt = 0.5 # time step
tmax = 30

plt.figure(figsize=(5,7))
plt.subplot(2,1,1)
for i in range(3):
    # simulation for ten seconds
    x = [0]  # initial position
    t = 0    # initial time
    while t < tmax:
        t += dt
        update(x)
    plt.plot(linspace(0,tmax,len(x)),x)
plt.yticks([-4,6])
plt.xlabel('time',fontsize=15)
plt.ylabel('position',fontsize=15)

plt.subplot(2,1,2)
plt.bar(velocities,probabilities,width=0.1,align='center')
plt.ylim([0,0.6])
plt.xticks(velocities)
plt.xlim([-0.8,0.8])
plt.xlabel('velocity',fontsize=15)
plt.ylabel('probability',fontsize=15)
plt.tight_layout()
plt.savefig('panelB.svg')
plt.show()
```

```python
def bounded_v(x):
    return [-0.2, 0.0, 0.2]

def bounded_probs(x):
    return [0.3*(1-x) + 0.4*x, 0.3, 0.4*(1-x) + 0.3*x]

def update_bounded(x):
    # Each compartment has its own probabilities and velocities
    index = int(round(x[-1])) # compartment 0 if x<0.5; compartment 1 if x>0.5 
    P = bounded_probs(x[-1])
    V = bounded_v(x[-1])
    
    # Update position based on local transport parameters 
    i = argwhere(random.multinomial(1,P))[0][0]
    new_x = x[-1]+V[i]*dt
    
    # Bound the position to be between zero and one
    if new_x > 1:
        x.append(1.0)
    elif new_x < 0:
        x.append(0.0)
    else:
        x.append(new_x)
```

```python
dt = 0.2 # time step
tmax = 90

plt.figure(figsize=(5,7))
plt.subplot(2,1,1)
for i in range(3):
    # simulation for ten seconds
    x = [0]  # initial position
    t = 0    # initial time
    while t < tmax:
        t += dt
        update_bounded(x)
    plt.plot(linspace(0,tmax,len(x)),x)
plt.yticks([0,1])
plt.xlabel('time',fontsize=15)
plt.ylabel('position',fontsize=15)

plt.subplot(2,1,2)
plt.bar(velocities[0],probabilities[0],width=0.1,align='center')
plt.ylim([0,0.6])
plt.xticks(velocities[0])
plt.xlim([-0.8,0.8])
plt.xlabel('velocity',fontsize=15)
plt.ylabel('probability',fontsize=15)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
for j in range(3):
    xx = []
    for i in range(100):
        # simulation for ten seconds
        x = [0.25] # initial position
        t = 0      # initial time
        while t < tmax:
            t += dt
            update_bounded(x)
        xx.append([round(x_) for x_ in x])
    xx = np.array(xx)
    tt = linspace(0,tmax,xx.shape[1])
    plt.plot(tt,np.mean(xx,axis=0),'-k',alpha=0.5)

plt.plot(tt,0.5 - 0.5*np.exp(-tt*0.04),'-r',lw=2)
```
